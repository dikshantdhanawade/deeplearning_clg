```python
import numpy as np

np.random.seed(42)

layers = 10

weight = []
biases = []


def relu(x):
    return np.maximum(0, x)

for i in range(layers):
    w = np.random.rand(1, 1)

    b = np.random.rand(1, 1)
    weight.append(w)
    biases.append(b)

def forward(x):
    for i in range(layers):
        x = np.dot(x, weight[i]) + biases[i]
        x = relu(x)
        print(x)
    return x


x = np.array([[1.0]])

output = forward(x)
print("output of DeepLearning model:", output)
```

Vanishing Gradient

```python
x = np.random.randn(1, 1)

weight_vanish = [np.random.rand(1, 1) * 0.01 for _ in range(layers)]

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


for w in weight_vanish:
    x = np.dot(x, w)
    print(x)
    x = sigmoid(x)
print("Final output",x)
```

Exploding Gradient

```python
weight_explode = [np.random.randn(1, 1) * 5 for _ in range(layers)]


for i, w in enumerate(weight_explode):
    x = np.dot(x, w)
    print(x)
    x = sigmoid(x)
print("final output:",x)
```
